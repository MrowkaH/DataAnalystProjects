```python
filename = "answers.txt"
raw = np.loadtxt(filename, delimiter=",")
data = raw[:,1:]
suma_uczniowie = np.sum(data, axis=1)
srednia = np.mean(suma_uczniowie)
mediana = np.median(suma_uczniowie)
print("podstawowe statystyki: \n")
print(f"srednia liczba poprawnych odpowiedzi: {srednia:.2f}")
print(f"mediana liczby poprawnych odpowiedzi: {mediana:.2f}")
odchylenie = np.std(suma_uczniowie)
cv = (odchylenie/srednia) * 100
print("\nzroznicowanie wynikow: \n")
print(f"odchylenie standardowe: {odchylenie:.2f}")
print(f"wspolczynnik zmienności: {cv:.2f}%")
n_pytan = data.shape[1]
procenty = (suma_uczniowie/n_pytan) * 100
zakresy = {
          "0% - 24%": np.sum((procenty>=0) & (procenty<25)),
          "25% - 49%": np.sum((procenty>=25) & (procenty<50)),
          "50% - 74%": np.sum((procenty>=50) & (procenty<75)),
          "75% - 100%": np.sum((procenty>=75) & (procenty<=100)),
            }
print("\nrozklad procentowy wynikow: \n")
for przedzial, liczba in zakresy.items():
  print(f"{przedzial}: {liczba} uczniow")
poprawne_na_pytanie = np.mean(data, axis=0) * 100
print("\nanaliza wynikow: \n")
for i, procent in enumerate(poprawne_na_pytanie, start=1):
  print(f"pytanie {i}: {procent:.1f}% uczniow odpowiedzialo poprawnie")
```

**Zadanie 2 — symulacja przewodnictwa ciepła w materiale (2D)**

Masz siatkę 2D (dwuwymiarową tablicę) reprezentującą temperatury na powierzchni prostokątnego materiału.  
Każda komórka siatki zawiera temperaturę w danym punkcie. Symulujemy przepływ ciepła przy pomocy iteracyjnej postaci równania przewodnictwa (relaksacja/Jacobi):

$$
T^{(n+1)}_{i,j} = \frac{1}{4}\left(T^{(n)}_{i+1,j} + T^{(n)}_{i-1,j} + T^{(n)}_{i,j+1} + T^{(n)}_{i,j-1}\right)
$$

Założenia:
- Stosujemy warunki brzegowe Dirichleta*(temperatury na krawędziach są stałe i nie ulegają aktualizacji podczas iteracji).
- Indeksowanie tablicy w opisie programistycznym jest *0-indeksowe* (pierwszy wiersz/kolumna mają indeks `0`).
- O ile nie zaznaczono inaczej, krawędzie niezmieniane w danym podpunkcie przyjmij jako stałe 0 °C.

1) Siatka $30 \times 30$
    - Inicjalizacja dowolna dla wnętrza (np. 0 °C).  
    - Warunki brzegowe:
      - górna krawędź: $100 \textdegree C$,
      - dolna krawędź: $0 \textdegree C$,
      - lewa i prawa krawędź: $0 \textdegree C$ (stałe przez całą symulację).
    - Wykonaj 50 kroków iteracji zgodnie ze wzorem powyżej (nie aktualizuj krawędzi).
    - Oblicz i podaj średnią temperaturę powierzchni po 25 oraz 50 krokach.
2) Siatka $100 \times 100$
    - Te same założenia jak w pkt 1 (góra $100 \textdegree C$, dół $0 \textdegree C$, boki $0 \textdegree C$, brzegi stałe).  
    - Wykonaj symulację i oblicz średnią temperaturę po 25 i 50 krokach.
3) Źródło ciepła wewnątrz — łatka $3 \times 3$
    - Ustaw całą powierzchnię (wnętrze i brzegi) początkowo np. na $t_s = 20 \textdegree C$.  
    - Wyznacz indeks startowy $s$ jako sumę trzech ostatnich cyfr Twojego numeru albumu.  
      - Przykład: numer albumu `012345` → $s = 3 + 4 + 5 = 12$.
    - Umieść w macierzy łatkę $3 \times 3$ o zadanej stałej temperaturze (np. wybierz i podaj wartość, np. $100 \textdegree C$) tak, aby lewy górny róg łatki był w wierszu `s` i kolumnie `s` (indeksowanie 0-indeksowe; w notacji 1-indeksowej byłby to 13. wiersz i 13. kolumna dla $s=12$).  
      - Jeśli $s > N-3$, przytnij $s$ do $N-3$, aby łatka mieściła się w siatce.
    - Traktuj tę łatkę jako stałe źródło (komórki łatki zachowują zadaną temperaturę podczas iteracji). Pozostałe brzegi pozostają stałe na $t_s$.
    - Po zakończeniu symulacji (np. po 50 krokach, chyba że prowadzący wskaże inaczej) oblicz średnią temperaturę w co 20. wierszu i co 20. kolumnie siatki (tj. w wierszach o indeksach `0, 20, 40, ...` oraz kolumnach `0, 20, 40, ...`). Zwróć te dwie średnie: osobno dla przekroju wierszowego i kolumnowego.



Wskazówki implementacyjne (`NumPy`):
- Aktualizuj tylko wnętrze siatki: `1:-1, 1:-1`.  
- Brzegi (i ewentualnie łatka $3 \times 3$) utrzymuj jako stałe między iteracjami.  
- Dla stabilności/zgodności z opisem używaj schematu Jacobi (czytaj wartości z poprzedniej macierzy, zapisuj do nowej).
- Po każdej iteracji możesz „nałożyć” warunki brzegowe/łatkę z powrotem (np. przez przypisanie do odpowiednich wycinków).

```python
def jacobi(T):
  T1 = T.copy()
  T1[1:-1, 1:-1] = 0.25 * (T[2:,1:-1] + T[:-2,1:-1] + T[1:-1,2:] + T[1:-1,:-2])
  return T1

N = 30
T = np.zeros((N, N))
T[0,:] = 100
T[-1,:] = 0
T[:,0] = 0
T[:,-1] = 0

for k in range(50):
  T = jacobi(T)
  T[0,:] = 100
  T[-1,:] = 0
  T[:,0] = 0
  T[:,-1] = 0
  if k + 1 in [25,50]:
    print(f"\nzadanie 1: srednia temperatura po {k+1} iteracjach = {np.mean(T):.2f}")
N = 100
T = np.zeros((N, N))
T[0,:] = 100
T[-1,:] = 0
T[:,0] = 0
T[:,-1] = 0

for k in range(50):
  T = jacobi(T)
  T[0,:] = 100
  T[-1,:] = 0
  T[:,0] = 0
  T[:,-1] = 0
  if k + 1 in [25, 50]:
    print(f"\nzadanie 2: srednia temperatura po {k+1} iteracjach = {np.mean(T):.2f}")
print("\nzadanie 3:")
N = 100
T = np.full((N, N), 20)
T[0,:] = 20
T[-1,:] = 20
T[:,0] = 20
T[:,-1] = 20
album = "160916"
s = sum(int(c) for c in album[-3:])
print(f"suma cyfr nr albumu = {s}")
n_s = s if s <= N - 3 else N - 3
T[n_s:n_s+3, n_s:n_s+3] = 100
for k in range(50):
  T_new = jacobi(T)
  T_new[0,:] = 20
  T_new[-1,:] = 20
  T_new[:,0] = 20
  T_new[:,-1] = 20
  T_new[n_s:n_s+3, n_s:n_s+3] = 100
  T = T_new
rows = np.arange(0, N, 20)
for r in rows:
  print(f"wiersz {r}: {np.mean(T[r,:]):.2f}")
for c in rows:
  print(f"kolumna {c}: {np.mean(T[:,c]):.2f}")
```

**Zadanie 3** (analiza danych pogodowych)

Stwórz fikcyjne, ale realistyczne dane pogodowe zebrane przez trzy stacje meteorologiczne.  
Każda stacja powinna być zapisana w osobnej macierzy `NumPy`.

Dane zbierane przez każdą stację obejmują:
- temperaturę $[^{\circ}C]$,
- wilgotność $[\%]$,
- opady $[mm]$.

Stacje rejestrują pomiary przez $7$ dni (wiersze) i $4$ razy dziennie (kolumny).

1. Oblicz średnią:
   - temperaturę,
   - wilgotność,
   - opady
   
   dla każdego dnia (średnia po czterech pomiarach) w każdej stacji.
2. Znajdź dzień oraz stację, w której zanotowano:
   - najwyższą temperaturę,
   - najniższe opady.
3. Sprawdź, czy wystąpiły dni z ekstremalnymi warunkami:
   - temperatura powyżej $30^{\circ}C$,
   - wilgotność powyżej $45\%$,
   - opady powyżej $20\,mm$.
   
   Jeśli takie dni wystąpiły, podaj, w której stacji i którego dnia.
4. Sprawdź, czy istnieją korelacje między temperaturą a wilgotnością  
   na wszystkich stacjach pomiarowych.  
   W tym celu użyj funkcji `np.corrcoef(x, y)` z biblioteki `NumPy`.

Wskazówki:
- Użyj osobnych macierzy dla każdej stacji lub jednej tablicy trójwymiarowej.
- Do uśredniania pomiarów w ciągu dnia możesz wykorzystać `np.mean(axis=1)`.
- Dane powinny być fikcyjne, ale realistyczne, np.:
  - temperatura: $10$–$35^{\circ}C$,
  - wilgotność: $30$–$90\%$,
  - opady: $0$–$25\,mm$.

```python
np.random.seed(42)
def generuj_dane():
  temperatura = np.random.uniform(10, 35, (7, 4))
  wilgotnosc = np.random.uniform(30, 90, (7, 4))
  opady = np.random.uniform(0, 25, (7, 4))
  return temperatura, wilgotnosc, opady
stacja1 = generuj_dane()
stacja2 = generuj_dane()
stacja3 = generuj_dane()
stacje = [stacja1, stacja2, stacja3]
print("srednie dzienne pomiary:\n")
for i, (temp, wilg, opad) in enumerate(stacje, start=1):
  srednia_temp = np.mean(temp, axis=1)
  srednia_wilg = np.mean(wilg, axis=1)
  srednia_opad = np.mean(opad, axis=1)
  print(f"stacja {i}:")
  for dzien in range(7):
    print(f" dzien {dzien+1}: temp={srednia_temp[dzien]:.1f}C, wilg={srednia_wilg[dzien]:.1f}%, opad={srednia_opad[dzien]:.1f}mm")
  print()
max_temp = -np.inf
min_opad = np.inf
for st_id, (temp, wilg, opad) in enumerate(stacje, start=1):
  dzien_max = np.argmax(np.mean(temp, axis=1))
  dzien_min_opad = np.argmin(np.mean(opad, axis=1))
  if np.mean(temp, axis=1)[dzien_max] > max_temp:
    max_temp = np.mean(temp, axis=1)[dzien_max]
    stacja_max_temp = st_id
    dzien_max_temp = dzien_max + 1
  if np.mean(opad, axis=1)[dzien_min_opad] < min_opad:
    min_opad = np.mean(opad, axis=1)[dzien_min_opad]
    stacja_min_opad = st_id
    dzien_min_opad = dzien_min_opad + 1
print(f"najwyzsza srednia temperatura: {max_temp:.1f}C w Stacji {stacja_max_temp}, Dzien {dzien_max_temp}")
print(f"najnizsze srednie opady: {min_opad:.1f}mm w Stacji {stacja_min_opad}, Dzien {dzien_min_opad}\n")
print("dni z ekstremalnymi warunkami:\n")
for st_id, (temp, wilg, opad) in enumerate(stacje, start=1):
  for dzien in range(7):
    t_srednia = np.mean(temp[dzien])
    w_srednia = np.mean(wilg[dzien])
    o_srednia = np.mean(opad[dzien])
    ekstremalne = []
    if t_srednia > 30:
      ekstremalne.append(f"temp>30C ({t_srednia:.1f}C)")
    if w_srednia > 45:
      ekstremalne.append(f"wilgotnosc>45% ({w_srednia:.1f}%)")
    if o_srednia > 20:
      ekstremalne.append(f"opady>20mm ({o_srednia:.1f}mm)")
    if ekstremalne:
      print(f"stacja {st_id}, dzien {dzien+1}: {'; '.join(ekstremalne)}")
print("\nkorelacja temperatura vs wilgotnosc:")
for st_id, (temp, wilg, _) in enumerate(stacje, start=1):
  temp_flat = temp.flatten()
  wilg_flat = wilg.flatten()
  corr = np.corrcoef(temp_flat, wilg_flat)[0,1]
  print(f"stacja {st_id}: korelacja = {corr:.2f}")
```
