**Projekt 1: Eksploracja danych oraz budowa i ocena modelu predykcyjnego (obowiązkowy)**

Projekt 1 jest projektem obowiązkowym i musi zostać wykonany łącznie z jednym wybranym projektem spośród Projektów 2–5.
Zakres tematyczny Projektu 1 nie może dublować zagadnień zrealizowanych w wybranym Projekcie 2–5.

Wykonaj następujące polecenia:
1. Wybierz samodzielnie zbiór danych i wczytaj go do programu (np. plik `CSV`, `Excel`).
2. Wykonaj eksplorację danych:
   - wyświetl podstawowe informacje o strukturze danych (`head()`, `info()`),
   - oblicz statystyki opisowe (`mean`, `median`, `std`),
   - przeanalizuj występowanie braków danych,
   - wykonaj wybrane wizualizacje danych (np. histogramy, wykresy punktowe, korelacje).
3. Przygotuj dane do dalszej analizy:
   - obsłuż braki danych (usunięcie lub imputacja),
   - wykonaj standaryzację lub normalizację cech,
   - zakoduj zmienne kategoryczne (jeśli występują),
   - podziel dane na zbiór treningowy i testowy.
4. Zbuduj model predykcyjny:
   - wybierz jeden typ zadania: klasyfikację lub regresję,
   - zaimplementuj jeden model predykcyjny,
   - krótko uzasadnij wybór modelu.
5. Oceń jakość modelu:
   - dla klasyfikacji oblicz `accuracy` oraz `F1-score`,
   - dla regresji oblicz `MSE` lub `RMSE`,
   - wykonaj co najmniej jedną wizualizację wyników
     (np. macierz pomyłek dla klasyfikacji lub wykres wartości rzeczywistych i przewidywanych dla regresji).
6. Sformułuj wnioski dotyczące jakości modelu, jego przydatności oraz ograniczeń zastosowanego podejścia.
7. Oddaj projekt w postaci:
   - kodu (skrypt `Python` lub notebook),
   - krótkiego raportu (`PDF` / `HTML` / `Markdown`).

**Projekt 2: Segmentacja klientów na podstawie danych Online Retail**

Dane: Online Retail ([Online Retail – UCI Machine Learning Repository](https://archive.ics.uci.edu/dataset/352/online+retail)).

Wykonaj następujące polecenia:
1. Wczytaj dane i wykonaj czyszczenie:
   - usuń rekordy z brakującymi wartościami w kluczowych polach (np. `CustomerID`, `InvoiceDate`, `Quantity`, `UnitPrice`),
   - usuń lub oznacz transakcje będące zwrotami (np. ujemne `Quantity` lub faktury/rekordy zwrotów).
2. Oblicz podstawowe statystyki dla wybranych zmiennych liczbowych (np. `Quantity`, `UnitPrice`, wartość koszyka):
   - średnia,
   - mediana,
   - odchylenie standardowe.
3. Utwórz cechy pomocnicze na poziomie transakcji:
   - wartość pozycji: `Quantity * UnitPrice`,
   - wartość zamówienia (Invoice): suma wartości pozycji w obrębie jednego `InvoiceNo`.
4. Zbuduj ramkę danych na poziomie klienta (jeden wiersz = jeden klient) zawierającą cechy RFM liczone względem daty `31.12.2011`:
   - `Recency`: liczba dni od ostatniej transakcji klienta do `31.12.2011`,
   - `Frequency`: liczba transakcji (np. liczba unikalnych `InvoiceNo`),
   - `Monetary`: całkowita wartość zakupów (suma wartości zamówień).
5. Na podstawie cech `Recency`, `Frequency` i `Monetary` utwórz trzy nowe kolumny:
   - `R_score`,
   - `F_score`,
   - `M_score`,
   z wartościami w skali `0–5` (`5` = najlepszy wynik).<br>
   Zasady punktacji:
   - `R_score`: mniejsza wartość `Recency` oznacza wyższą ocenę,
   - `F_score` oraz `M_score`: większe wartości oznaczają wyższą ocenę.
6. Wykonaj segmentację klientów na podstawie cech `Recency`, `Frequency` i `Monetary` (po wcześniejszej standaryzacji danych):
   - przygotuj dane do klasteryzacji (np. standaryzacja),
   - wykonaj klasteryzację (np. `KMeans`; dopuszczalnie `DBSCAN`),
   - wybierz liczbę klastrów (np. metoda łokcia lub metoda profilu).
7. Opisz otrzymane grupy klientów:
   - podaj średnie wartości cech (`R`, `F`, `M` oraz/lub `Recency`, `Frequency`, `Monetary`) w każdym klastrze,
   - krótko nazwij segmenty (np. „lojalni”, „wysokowartościowi”, „uśpieni” itp.).
8. Zwizualizuj segmenty w zredukowanej wymiarowości:
   - zastosuj `PCA` (lub `t-SNE`) na danych użytych do klasteryzacji,
   - wykonaj wykres 2D z kolorami odpowiadającymi klastrom.
9. Sprawdź, który segment ma największy udział zwrotów:
   - zdefiniuj zwrot (np. ujemne `Quantity`),
   - policz liczbę lub udział zwrotów dla każdego segmentu,
   - przedstaw wynik w tabeli lub na wykresie.
10. Oddaj projekt w postaci:
    - kodu (skrypt Python lub notebook),
    - krótkiego raportu (`PDF`/`HTML`/`Markdown`) z wykresami i wnioskami.

**Projekt 3: Analiza danych i modelowanie regresyjne z obsługą wartości odstających**

Dane: `data.csv` (plik udostępniony przez prowadzącego).

Wykonaj następujące polecenia:

   - uzupełnij braki w kolumnach kategorycznych (np. najczęstszą wartością) lub usuń wiersze z brakami.
2. Zbadaj rozkład zmiennej `Income`:

   - wykonaj wykres pudełkowy.
3. Zidentyfikuj wartości odstające w `Income` metodą IQR:
   - wyznacz dolną i górną granicę IQR,
   - oznacz obserwacje odstające (np. kolumna `Income_Outlier` z wartościami `0/1`),
   - porównaj podstawowe statystyki `Income` przed i po usunięciu odstających (lub po przycięciu wartości).

   - model 1: regresja liniowa na danych bez dodatkowych operacji na odstających,
   - model 2: regresja liniowa na danych po obsłudze odstających (usunięcie/przycięcie lub transformacja `log1p` zmiennej `Income`).
   - porównaj wyniki modeli metrykami `RMSE` oraz `R2`.
6. Podsumuj wyniki:
   - opisz, jak wartości odstające wpływały na `Income` i jakość modelu,
   - wskaż, który wariant obsługi odstających dał lepsze wyniki i dlaczego.
