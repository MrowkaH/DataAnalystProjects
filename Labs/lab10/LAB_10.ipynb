**Zadanie 2**
Wykorzystaj PCA (Analizę Głównych Składowych) do redukcji wymiarowości danych z zadania 1 oraz dokonaj interpretacji otrzymanych głównych składowych.
Przed zastosowaniem PCA wykonaj standaryzację danych przy użyciu klasy `StandardScaler`.

W szczególności:
- usuń kolumnę `Wine`, traktując ją jako etykietę (nie biorącą udziału w PCA),
- wykonaj standaryzację cech numerycznych,
- przeprowadź PCA i sprawdź, ile wariancji wyjaśniają kolejne główne składowe,
- zinterpretuj znaczenie pierwszych głównych składowych na podstawie wag (ładunków) cech.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# 1. Usunięcie kolumny Wine
X = df_wine.drop('Wine', axis=1)
y = df_wine['Wine']

# 2. Standaryzacja
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. PCA
pca = PCA() # Domyślnie tyle składowych ile cech
X_pca = pca.fit_transform(X_scaled)

# Wyjaśniona wariancja
explained_variance = pca.explained_variance_ratio_
print("Wyjaśniona wariancja przez poszczególne składowe:")
print(explained_variance)
print(f"\nŁączna wariancja wyjaśniona dla 2 pierwszych składowych: {sum(explained_variance[:2]):.4f}")

# Wykres osypiska (Scree plot)
plt.figure(figsize=(8, 4))
plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o', linestyle='--')
plt.title('Wykres osypiska (Scree Plot)')
plt.xlabel('Liczba składowych')
plt.ylabel('Wyjaśniona wariancja')
plt.grid()
plt.show()

# Interpretacja wag (ładunków) dla pierwszych 2 składowych
components = pd.DataFrame(pca.components_, columns=X.columns, index=[f'PC{i+1}' for i in range(len(X.columns))])
print("\nWagi (ładunki) dla pierwszych 2 składowych:")
display(components.iloc[:2].T)
```

**Zadanie 3**
1. Podziel dane z zadania 1 na skupienia, wykorzystując do obliczeń składowe otrzymane w zadaniu 2 (PCA), za pomocą dwóch algorytmów:
   - KMeans: algorytm centroidów.
   - DBSCAN: gęstościowa analiza skupień.
2. Oceń jakość klasteryzacji za pomocą wskaźników:
   - wewnętrznych wskaźników jakości:
     - współczynnik sylwetki (Silhouette),
     - indeks Daviesa-Bouldina (DBI);
   - zewnętrznego wskaźnika jakości (z wykorzystaniem etykiet `Wine`):
     - skorygowany współczynnik Randa (ARI).
3. Porównaj wyniki uzyskane z dwóch algorytmów i omów różnice (liczba skupień, jakość wg wskaźników, wpływ szumu w DBSCAN).

```python
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score, adjusted_rand_score

X_pca_2 = X_pca[:, :2]

# 1. KMeans
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels_kmeans = kmeans.fit_predict(X_pca_2)

# 2. DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
labels_dbscan = dbscan.fit_predict(X_pca_2)

# Funkcja do oceny
def evaluate_clustering(algorithms_name, labels, true_labels, data):
    print(f"--- {algorithms_name} ---")
    
    unique_labels = set(labels)
    if len(unique_labels) < 2:
        print("Mniej niż 2 klastry - niemożliwa ocena wskaźnikami wewnętrznymi.")
        return

    sil = silhouette_score(data, labels)
    dbi = davies_bouldin_score(data, labels)
    ari = adjusted_rand_score(true_labels, labels)

    print(f"Silhouette Score: {sil:.4f}")
    print(f"Davies-Bouldin Index: {dbi:.4f}")
    print(f"Adjusted Rand Index (ARI): {ari:.4f}")
    print(f"Liczba klastrów: {len(unique_labels) - (1 if -1 in unique_labels else 0)}")
    print(f"Liczba punktów szumu: {list(labels).count(-1)}")

evaluate_clustering("KMeans", labels_kmeans, y, X_pca_2)
evaluate_clustering("DBSCAN", labels_dbscan, y, X_pca_2)

# Porównanie wizualne
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.scatterplot(x=X_pca_2[:, 0], y=X_pca_2[:, 1], hue=labels_kmeans, palette='viridis', ax=axes[0])
axes[0].set_title('KMeans (na 2 składowych PCA)')
sns.scatterplot(x=X_pca_2[:, 0], y=X_pca_2[:, 1], hue=labels_dbscan, palette='viridis', ax=axes[1])
axes[1].set_title('DBSCAN (na 2 składowych PCA)')
plt.show()
```

**Zadanie 4**
Dla danych z zadania 1 oraz składowych głównych otrzymanych w zadaniu 2:
1. Wygeneruj dendrogram przedstawiający hierarchiczne łączenie skupień.
2. Na podstawie dendrogramu wybierz optymalną liczbę skupień.
3. Podziel dane na skupienia za pomocą aglomeracyjnej analizy skupień.
4. Zwizualizuj otrzymane wyniki klasteryzacji.

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering

# 1. Dendrogram
plt.figure(figsize=(10, 5))
Z = linkage(X_pca_2, method='ward')
dendrogram(Z)
plt.title('Dendrogram (Ward)')
plt.xlabel('Próbki')
plt.ylabel('Odległość euklidesowa')
plt.show()

# 2. Wybór liczby skupień i klasteryzacja
n_clusters_agg = 3
agg_cl = AgglomerativeClustering(n_clusters=n_clusters_agg, metric='euclidean', linkage='ward')
labels_agg = agg_cl.fit_predict(X_pca_2)

# 3. Wizualizacja
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca_2[:, 0], y=X_pca_2[:, 1], hue=labels_agg, palette='viridis')
plt.title(f'Aglomeracyjna analiza skupień (n={n_clusters_agg})')
plt.show()

# Ocena
evaluate_clustering("Agglomerative Clustering", labels_agg, y, X_pca_2)
```

**Zadanie 6**
Wczytaj dane pacjentów z pliku `patients.csv` i wykonaj następujące kroki:
1. Przygotuj dane do klasteryzacji.
   Do obliczeń nie wykorzystuj kolumny `Group` (traktuj ją jako informację pomocniczą).
2. Przeprowadź klasteryzację pacjentów algorytmem `KMeans` (inicjalizacja kmeans++).
   Dobierz liczbę skupień z wykorzystaniem:
   - metody łokcia,
   - metody profilu.
3. Przeprowadź klasteryzację pacjentów z wykorzystaniem hierarchicznego grupowania (aglomeracyjnej analizy skupień).
4. Oceń jakość klastrów za pomocą metody profilu.
   Następnie przeanalizuj, jak różnią się grupy pacjentów, porównując średnie wartości cech w poszczególnych klastrach.
5. Zwizualizuj wyniki obu algorytmów w zredukowanej wymiarowości, wykorzystując analizę głównych składowych (PCA).

```python
# 1. Wczytanie i przygotowanie danych
df_patients = pd.read_csv('patients.csv')
X_pat = df_patients.drop('Group', axis=1)
y_pat = df_patients['Group']

# Standaryzacja
scaler_pat = StandardScaler()
X_pat_scaled = scaler_pat.fit_transform(X_pat)

# 2. Dobór liczby skupień
inertia = []
sil_scores = []
K_range = range(2, 11)

for k in K_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X_pat_scaled)
    inertia.append(km.inertia_)
    sil_scores.append(silhouette_score(X_pat_scaled, km.labels_))

# Wykresy pomocnicze
fig, ax = plt.subplots(1, 2, figsize=(14, 5))
ax[0].plot(K_range, inertia, marker='o')
ax[0].set_title('Metoda łokcia (Elbow Method)')
ax[0].set_xlabel('Liczba klastrów k')
ax[0].set_ylabel('Inercja')
ax[0].grid()

ax[1].plot(K_range, sil_scores, marker='o', color='orange')
ax[1].set_title('Metoda profilu (Silhouette Score)')
ax[1].set_xlabel('Liczba klastrów k')
ax[1].set_ylabel('Silhouette Score')
ax[1].grid()
plt.show()

k_opt = 3 
print(f"Wybrano liczbę klastrów: {k_opt}")

# 3. Klasteryzacja KMeans
kmeans_pat = KMeans(n_clusters=k_opt, init='k-means++', random_state=42, n_init=10)
labels_kmeans_pat = kmeans_pat.fit_predict(X_pat_scaled)

# 4. Klasteryzacja Hierarchiczna
agg_pat = AgglomerativeClustering(n_clusters=k_opt)
labels_agg_pat = agg_pat.fit_predict(X_pat_scaled)

# 5. Ocena jakości
print(f"Silhouette Score (KMeans, k={k_opt}): {silhouette_score(X_pat_scaled, labels_kmeans_pat):.4f}")
print(f"Silhouette Score (Agglomerative, k={k_opt}): {silhouette_score(X_pat_scaled, labels_agg_pat):.4f}")

# Analiza różnic grup (średnie cech)
df_analysis = X_pat.copy()
df_analysis['Cluster_KMeans'] = labels_kmeans_pat
print("\nŚrednie wartości cech w klastrach (KMeans):")
display(df_analysis.groupby('Cluster_KMeans').mean())

# 6. Wizualizacja PCA
pca_pat = PCA(n_components=2)
X_pat_pca = pca_pat.fit_transform(X_pat_scaled)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# KMeans PCA
sns.scatterplot(x=X_pat_pca[:, 0], y=X_pat_pca[:, 1], hue=labels_kmeans_pat, palette='viridis', ax=axes[0], s=60)
axes[0].set_title(f'KMeans (k={k_opt}) - rzut na PCA')
axes[0].set_xlabel('PC1')
axes[0].set_ylabel('PC2')

# Agglomerative PCA
sns.scatterplot(x=X_pat_pca[:, 0], y=X_pat_pca[:, 1], hue=labels_agg_pat, palette='viridis', ax=axes[1], s=60)
axes[1].set_title(f'Agglomerative (k={k_opt}) - rzut na PCA')
axes[1].set_xlabel('PC1')
axes[1].set_ylabel('PC2')

plt.show()
```
